```python
import cv2

import matplotlib.pyplot as plt

import numpy as np

import pandas as pd
```

### Using PIL and Image

```python
### Reading and Displaying using PIL

from PIL import Image

pic = Image.open('./Section2Images/Image1.jpg')
```

```python
### If we use Image we need to convert it into a numpy array

pil_img = np.array(pic)

print(pil_img.shape)

plt.imshow(pil_img)
```

### Using CV2

```python
img = cv2.imread('./Section2Images/Image1.jpg')

plt.imshow(img)
```

```python
B, G, R = img[10, 50] # BGR Values for Pixel(10,50)

print(img.shape)

print(B, G, R)
```

### Converting to GrayScale

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray_img , cmap = 'gray')
```

```python
print(gray_img.shape)

print(gray_img[10, 50])
```

### Converting to HSV

```python
#H: 0 - 180, S: 0 - 255, V: 0 - 255

hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

plt.imshow(hsv_img)
```

#### Hue Channel

```python
plt.imshow(hsv_img[:, :, 0])
```

#### Saturation Channel

```python
plt.imshow(hsv_img[:, :, 1])
```

####  Value Channel

```python
plt.imshow(hsv_img[:, :, 2])
```
