```python
import cv2

import matplotlib.pyplot as plt

import numpy as np

import pandas as pd
```

```python
img = cv2.imread('./Section2Images/Image1.jpg')

cv2.imshow('Original', img)

cv2.waitKey()

cv2.destroyAllWindows()
```

```python
plt.imshow(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2.imshow('Grayscale', img)

cv2.waitKey()

cv2.destroyAllWindows()
```

```python
cv2.imwrite('./Section2Images/Section2Image1InGrayscale.png', img)
```
