```python
import cv2

import numpy as np

import pandas as pd

import matplotlib.pyplot as plt
```

```python
image = cv2.imread('./Section3Images/Image1.jpg')

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

smaller = cv2.pyrDown(image)

larger = cv2.pyrUp(smaller)
```

## Normal

```python
plt.imshow(image)
```

## Smaller

```python
plt.imshow(smaller)
```

## Larger

```python
plt.imshow(larger)
```
