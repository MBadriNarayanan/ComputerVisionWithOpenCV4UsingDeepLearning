```python
import matplotlib.pyplot as plt

import numpy as np

import pandas as pd

import cv2
```

## SIFT

```python
image = cv2.imread('./Section5Images/Image1.jpg')

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

rgbimage = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(gray , cmap = "gray")
```

```python
#Create SIFT Feature Detector object

sift = cv2.xfeatures2d.SIFT_create()

#Detect key points

keypoints = sift.detect(rgbimage, None)

print(" Number of keypoints Detected: ", len(keypoints))
```

```python
rgbimage = cv2.drawKeypoints(rgbimage, keypoints, flags = cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS , outImage = None)

plt.imshow(rgbimage)
```

## SURF

```python
image = cv2.imread('./Section5Images/Image1.jpg')

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

rgbimage = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(gray , cmap = "gray")
```

```python
surf = cv2.xfeatures2d.SURF_create()

# Only features, whose hessian is larger than hessianThreshold are retained by the detector

surf.setHessianThreshold(1000)

keypoints, descriptors = surf.detectAndCompute(rgbimage, None)

print(" Number of keypoints Detected: ", len(keypoints))
```

```python
rgbimage = cv2.drawKeypoints(rgbimage, keypoints, flags = cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS, outImage = None)

plt.imshow(rgbimage)
```

## FAST

```python
image = cv2.imread('./Section5Images/Image1.jpg')

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

rgbimage = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(gray , cmap = "gray")
```

```python
fast = cv2.FastFeatureDetector_create()

# Obtain Key points, by default non max suppression is On
# to turn off set fast.setBool('nonmaxSuppression', False)
keypoints = fast.detect(rgbimage, None)
print(" Number of keypoints Detected: ", len(keypoints))
```

```python
rgbimage = cv2.drawKeypoints(rgbimage, keypoints, flags = cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS, outImage = None)

plt.imshow(rgbimage)
```

## BRIEF

```python
image = cv2.imread('./Section5Images/Image1.jpg')

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

rgbimage = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(gray , cmap = "gray")
```

```python
fast = cv2.FastFeatureDetector_create()

# Create BRIEF extractor object

brief =  cv2.xfeatures2d.BriefDescriptorExtractor_create()

# Determine key points

keypoints = fast.detect(rgbimage , None)

# Obtain descriptors and new final keypoints using BRIEF

keypoints, descriptors = brief.compute(rgbimage , keypoints)

print(" Number of keypoints Detected: ", len(keypoints))
```

```python
rgbimage = cv2.drawKeypoints(rgbimage, keypoints, flags = cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS, outImage = None)

plt.imshow(rgbimage)
```

## ORB

```python
image = cv2.imread('./Section5Images/Image1.jpg')

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

rgbimage = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(gray , cmap = "gray")
```

```python
orb = cv2.ORB_create()

# Determine key points

keypoints = orb.detect(rgbimage, None)

# Obtain the descriptors

keypoints, descriptors = orb.compute(rgbimage, keypoints)

print(" Number of keypoints Detected : ", len(keypoints))
```

```python
rgbimage = cv2.drawKeypoints(rgbimage, keypoints, flags = cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS, outImage = None)

plt.imshow(rgbimage)
```
